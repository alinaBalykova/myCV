## Задача 3

### **3.1 Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.**

1. Анализ взаимных распределений по показателю цены и оценки жилья

```python
sns.jointplot(x=df['grade'], y=df['price'], kind='reg', height=12, ratio=4)
plt.xlabel('Класс')
plt.ylabel('Цена')
plt.title('Распределение цен в зависимости от класса дома');
```

Boxplot вариант

```python
plt.figure(figsize=(14, 10))

sns.boxplot(x=df['grade'], y=df['price'], whis=1.0)

plt.xlabel('Класс')
plt.ylabel('Цена')
plt.title('Распределение цен в зависимости от класса дома');
```

Вывод: Можно заметить, что увеличение уровня класса (grade) влияет на рост цен недвижимости (экспоненциальное распределение?); для домов с 7-11 классом наблюдается максимальное количество выбросов, что может быть связано с влиянием других характеристик жилья на ценообразование

2. Анализ влияния наличия вида на набережную на цену недвижимости

```python
plt.figure(figsize=(14, 10))
sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5, hue=df['grade'])

plt.title('Распределение цен')
plt.xlabel('Цена')
plt.ylabel('Наличие набережной');
```

Вывод: Медиана цен домов, не имеющих вид на набережную меньше медианы цен домов, имеющих вид на набережную
Разница цены для домов с набережной больше, чем для домов без нее. Скорее всего это связано с разницей с другими характеристиками жилья: в качестве ремонта, кв.м, реновации итд

3. Анализ количества комнат и ванных комнат на ценообразование

```python
plt.figure(figsize=(15, 6))
plt.scatter(df['bedrooms'], df['price'])
plt.title('Распределение средних цен в зависимости от количества спален')
plt.ylabel('Цена')
plt.xlabel('Количество спален');
```

```python
sns.jointplot(data=df, x="bedrooms", y="price", kind="reg", height=12, ratio=4);
```

```python
sns.jointplot(data=df, x="bathrooms", y="price", kind="reg", height=12, ratio=4);
```

Вывод: Цена зависит от количества комнат. Судя по графику - распределение близко к нормальному. Чаще покупают от 1-5 комнат.
Также цена зависит количества ванных комнат, что отражает положительная корреляция.

4. В целом посмотрим матрицу корреляций

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10, 8))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

Заметная корреляция у цены с характеристиками : sqft_living, grade, sqft_above,sqft_living15, bathrooms

5. Посмотрим зависимость цены от года постойки

```python
data = df.groupby('yr_built')['price'].mean().reset_index()
data
```

```python
plt.figure(figsize=(15, 6))
plt.scatter(data['yr_built'], data['price'])
plt.title('Распределение средних цен в зависимости от года постройки')
plt.ylabel('Цена')
plt.xlabel('Год постройки');
```

 распределение цены в зависимости от года постройки имеет U-образную форму; Начиная с 1980 год виден подъем цен. Высокие цены с 1900 по 1920 скорее всего связаны с исторической ценностью здания. С 1940 по 1980 виден спад цен, что возможно свзяно с последствиями Второй мировой.

Общий вывод:

В целом, можно заключить, что ценообразование зависит от нескольких параметров: количества комнат (в том числе ванных), от приближенности к набережной, от года постройки и от квадратных метров.
